```python
#Especially used in Machine Learning, Statistics
#Also called as Indicator Variables
#Suppose you have a column called Sex
#A dummy variablewill hold Female as 0, Male as 1
import pandas as pd
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
#lets use series map method to create dummy variable
train['sex_male'] = train.Sex.map({'female':0, 'male':1})
#here creating a new colmn called sex_male
train.head()
```

```python
#there is another flexible way of doing the above
pd.get_dummies(train.Sex)
#what this does, it creates column each for both Male and Female
#and on that row, if it mentions as Male,this new created column would be 0 1, where 1 is the Male Column
#and so on
#since we have two categorical values here, therefore we would use one dummy variable
#remember the code: k - 1 to have dummy variable
#Since i dont want the first column here, that says female, lets drop the column female
```

```python
pd.get_dummies(train.Sex).iloc[:,1:]
#here i want all the rows,
#and i want the second column, since 0th column would be 1
```

```python
#lets prefix with Sex
#samecode as of above
pd.get_dummies(train.Sex, prefix= 'Sex').iloc[:,1:]
```

```python
#just for our understanding, lets create dummy vairables for 'Embarked'
#the embarkedcolumn has 3 categoricalvalues, C, Q, S
pd.get_dummies(train.Embarked, prefix = 'Embarked')
#we will have three columns with C Q S
#whereever row of S is present, that column would be marked as 1 and so on for other categoricalvalues
```

```python
#but to have dummy variables, we need k - 1
```

```python
pd.get_dummies(train.Embarked, prefix = 'Embarked').iloc[:,1:]
```

```python
#embarked values: 0,1 => thats S
#Embarked values: 0,0 => thats C
#EmbarkedValues, 1,0 => thats Q
#nowlets attach this to the oru=iginal data set

emb_dummies = pd.get_dummies(train.Embarked, prefix = 'Embarked').iloc[:,1:]
```

```python
#lets concatenate
train = pd.concat([train, emb_dummies], axis= 1)
#if you want to concatentae rows, use axis as 0, from topto bottom, row wise
#if you want to concatenate columns, use axis as 1, from left to right, column wise
#here new column Sex_male is also added
```

```python
#iloc, loc, ix
#these are different rows and columns for selecting rows and columns
import pandas as pd
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head(3)
```

```python
#lets use loc: 
#used for filtering rows and columns by label
#by labels i mean, rows label = index
#by labels, column = column names
#loc: what rows i want and what columns i want
#loc[ , ]
 
```

```python
#eg: first row and all columns
ufo.loc[0, :]
```

```python
#for multiple rows
ufo.loc[[0,1,2], :]
```

```python
#above code can be rewritten as:
ufo.loc[0:2 , :]
```

```python
#Above has been for row selections
#lets look into column selections
ufo.loc[: , 'City']
```

```python
#for multiple columns
ufo.loc[:, ['City', 'State']]
```

```python
#again for multiple columns
ufo.loc[:, 'City':'State']
```

```python
#another way of writing above is:
ufo.head(3).drop('Time', axis = 1)
```

```python
#Selecting column with Oakland
ufo[ufo.City == 'Oakland']
```

```python
#above code can be written as in loc
ufo.loc[ufo.City == 'Oakland', :]
```

```python
#iloc
#selecting rows and columns by integer position
ufo.iloc[ : , [0,3]]
#if we used range: 0:4
#result will be 0,1,2,3
#iloc is exclusive of 2nd element
#loc is inclusive of 1st element
```

```python
#Another example
ufo.iloc[0:3, :]
#here it will not include row 3,
#but in loc will include
```

```python
#loc: for labels, and inclusive of both sides
#iloc: for integer positions, and inclusive of start and exclusive of stop
```

```python
#sometimes you get codes as
ufo[0:2]
#this will result into row selection
```

```python
#ix
#allows you to mix labels and integers when doing selections
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col = 'country')
drinks.head()
```

```python
#from table i want to select 89,
#here label is index name
#integer wil be col
drinks.ix['Albania', 0]
```

```python
#another example
drinks.ix[1, 'beer_servings']
#here 1 is the integer position, and beer_servings is the label 
```

```python
#another example
drinks.ix['Albania':'Andorra', 0:3]
```

```python
#logic in ix is: if you have stirngs and numbers
#strings are considered as labels and numbers are positions
#in iloc, its inclusive of position 0 and exclusive of position 2
```

```python
#another example
ufo.ix[0:2, 0:2]
```

```python
#here it gave me 3 rows and 2 columns
#u have a string index with columns and give it numbers, it will do it by position
#if you have integer index, and give it numbers, it will treat as labels
#and when using labels, its inclusive on both sides
#always try to avoid ix
#better to use loc and iloc
```

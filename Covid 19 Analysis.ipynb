```python
from sklearn.impute import SimpleImputer
#to handle missing data, we could have used fillna function in python
```

```python
#remove SrNo and LastUodate
df.drop(['SNo', 'Last Update'], axis = 1, inplace = True)#axis1 means colns,
```

```python
#Rename all colns
df.rename(columns = {'ObservationDate': 'Date', 'Province/State': 'Province', 'Country/Region': 'Country'}, inplace = True)
```

```python
#change date coln to pandas dateformat
df['Date'] = pd.to_datetime(df['Date'])
```

```python
#we see, province entries are with null values, all other colns are with 38622
#province has got empty records, hence we need to convert them to NA first
df = df.fillna('NA')#we cant use inplace here, we need to save it in a new variable
```

```python
df.info()
```

```python
#now we dont see any empty values in province
```

```python
imputer = SimpleImputer(strategy = 'constant')
df2 = pd.DataFrame(imputer.fit.transform(df), columns = df.columns)
```

```python
#lets groupby
#how many total cases are there in each country
df2 = df.groupby('Country')[['Country', 'Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
#grouping by Country, and i want to see values of Country, COnfirmed....
#after this, i want to get all the sums of those values under each country
```

```python
df2
```

```python
#Grouping Country and Date
df2 = df.groupby(['Country', 'Date'])[['Country', 'Date', 'Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

```python
df2.head(10)
```

```python
#all records, for confirmed cases that are more than 100
df3 = df2[df2['Confirmed'] > 100]
```

```python
df3
```

```python
import matplotlib.pyplot as plt
```

```python
#lets check how many countries are there in total
countries = df3['Country'].unique()
len(countries)
#all the countries without repetation
```

```python
#my goal for each country i want to see what is the trend with respect to date, trend for death patients,
#recovered patients as date moves on
import numpy as np
for idx in range(0, len(countries)):
    C = df3[df3['Country'] == countries[idx]].reset_index()
    plt.scatter(np.arange(0,len(C)), C['Confirmed'], color = 'blue', label = 'Confirmed')
    plt.scatter(np.arange(0,len(C)), C['Recovered'], color = 'green', label = 'Recovered')
    plt.scatter(np.arange(0,len(C)), C['Deaths'], color = 'red', label = 'Deaths')
    plt.title(countries[idx])
    plt.xlabel('Days since the first suspect')
    plt.ylabel('Number of Cases')
    plt.legend()
    plt.show()
```

```python
#trends for all the world together
df4 = df3.groupby(['Date'])[['Date', 'Confirmed', 'Deaths', 'Recovered']].sum().reset_index()
```

```python
C = df4
plt.scatter(np.arange(0,len(C)), C['Confirmed'], color = 'blue', label = 'Confirmed')
plt.scatter(np.arange(0,len(C)), C['Recovered'], color = 'green', label = 'Recovered')
plt.scatter(np.arange(0,len(C)), C['Deaths'], color = 'red', label = 'Deaths')
plt.title('World')
plt.xlabel('Days Since the First Suspect')
plt.ylabel('Number of Cases')
plt.legend()
plt.show()
```
